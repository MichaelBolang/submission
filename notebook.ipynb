## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana statistik rental sepeda di tahun 2012?

```python
df = set_day.query('yr == 1')
statistik_2012 = df.groupby('mnth')['cnt'].sum().reset_index()

# Visualisasi data dengan line chart
plt.figure(figsize=(10, 6))
sn.lineplot(data=statistik_2012, x='mnth', y='cnt')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Sewa Sepeda')
plt.title('Statistik Rental Sepeda pada Tahun 2012')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Agu', 'Sep', 'Okt', 'Nov', 'Des'])
plt.show()
```

- Berdasarkan line chart diatas, bisa terlihat bahwa jumlah rental sepeda pada bulan Januari memiliki jumlah paling kecil yang kemudian akan terus naik jumlahnya hingga puncaknya pada bulan September. Kemudian, akan terjadi penurunan dari bulan September menuju bulan Desember. Hal ini kemungkinan besar terjadi karena temperatur memasuki bulan September merupakan cuaca dan temperatur yang cocok untuk bersepeda.

### Pertanyaan 2: Apa musim yang paling banyak terjadi rental sepeda di tahun 2011?

```python
df = set_day.query('yr == 0')
hasil_2011 = df.groupby('season')['cnt'].sum().reset_index()

# Visualisasi data
plt.figure(figsize=(10, 6))
plt.bar(hasil_2011['season'], hasil_2011['cnt'])
plt.xlabel('Musim')
plt.ylabel('Jumlah Sewa Sepeda')
plt.title('Jumlah Sewa Sepeda Berdasarkan Musim pada Tahun 2011')
plt.xticks(range(1, 5), ['Spring', 'Summer', 'Fall', 'Winter'])
plt.show()
```

- Berdasarkan bar plot diatas, bisa terlihat bahwa jumlah rental sepeda terbanyak terdapat pada musim Fall. Hal ini kemungkinan besar terjadi karena cuaca dan temperatur pada musim fall merupakan yang paling cocok untuk melakukan aktivitas bersepeda.

## Conclusion

- Conclution pertanyaan 1 : Bulan Januari memiliki jumlah rental sepeda terkecil yang akan mengalami penaikan hingga puncaknya berada di bulan September yang setelah itu akan terjadi penurunan hingga bulan Desember pada tahun 2012
- Conclution pertanyaan 2 : Musim Fall memiliki jumlah rental sepeda terbanyak pada tahun 2011 diikuti oleh musim Summer di tempat kedua, musim Winter di tempat ketiga, dan musim Spring di tempat terakhir
